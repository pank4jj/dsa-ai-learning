```python
class MyRectangle:
    def __init__(self, length, breadth):
        self.length = length
        self.breadth = breadth
        self.random_attribute = 0
        self.another_random_attribute = 'abc'

    def get_perimeter(self):
        return 2 * (self.length + self.breadth)

    def get_area(self):
        return self.length * self.breadth

    def get_length(self):
        return self.length

    def get_breadth(self):
        return self.breadth
```

```python
obj1 = MyRectangle(length=5, breadth=6)
```

```python
obj2 = MyRectangle(length=10, breadth=12)
```
